**Explanation:**

**Example:**

* **Image recognition systems** that can identify objects in photos (like Facebook tagging friends automatically) use deep learning models to understand complex visual patterns.


Q2: What is supervised learning? List some examples of supervised learning.

ans.
### Supervised Learning?

Supervised learning is a type of machine learning where the model is trained on a labeled dataset. This means each training example includes both the input data and the correct output (label). The goal is for the model to learn a mapping from inputs to outputs so it can predict the output for new, unseen data.

**In simple terms:**
The model learns from examples where the answers are already known, like a student learning with the answer key.


### Examples of Supervised Learning:

1. **Email Spam Detection**

* Input: Email content
* Output: Label indicating “spam” or “not spam”
* The model learns to classify emails based on labeled examples.

2. **Image Classification**

* Input: Images of animals
* Output: Labels like “cat,” “dog,” “bird,” etc.
* The model learns to identify the animal in a new image.

3. **House Price Prediction**

* Input: Features like size, location, number of bedrooms
* Output: Actual house price
* The model learns to predict house prices based on past sales data.

4. **Sentiment Analysis**

* Input: Text from reviews or social media posts
* Output: Sentiment label such as “positive,” “neutral,” or “negative”
* The model learns to detect sentiment from labeled text.


Q3:  What is unsupervised learning? List some examples of unsupervised learning.

Ans.

###Unsupervised Learning?

Unsupervised learning is a type of machine learning where the model is given data **without any labels** or predefined outcomes. The goal is to find hidden patterns, groupings, or structures within the data without any guidance on what the results should be.

**In simple terms:**
The model tries to learn the underlying structure or relationships in the data on its own, like exploring without a map.


### Examples of Unsupervised Learning:

1. **Clustering**

* Task: Group similar data points together based on features.
* Example: Customer segmentation in marketing — grouping customers based on their buying behavior without knowing their categories in advance.

2. **Anomaly Detection**

* Task: Identify unusual data points that don’t fit the normal pattern.
* Example: Detecting fraudulent transactions in banking where unusual patterns indicate fraud.

3. **Dimensionality Reduction**

* Task: Reduce the number of variables while preserving important information.
* Example: Compressing large image datasets for easier processing using techniques like Principal Component Analysis (PCA).

4. **Association Rule Learning**

* Task: Find interesting relationships between variables in large databases.
* Example: Market basket analysis — figuring out which products are often bought together by customers.

Q6: What is train, test and validation split? Explain the importance of each term.

ans.  Train, Test, and Validation Split?

When building a machine learning model, the available data is usually divided into three parts:

1. **Training Set**

* This is the portion of data used to **train the model** — the model learns patterns, relationships, and features from this data.
* Usually the largest portion (e.g., 60-80% of total data).

2. **Validation Set**

* This set is used to **tune the model’s hyperparameters** (like learning rate, number of layers, etc.) and to monitor performance during training.
* Helps to **avoid overfitting** by checking how well the model generalizes on unseen data during the training process.
* Typically around 10-20% of the data.

3. **Test Set**

* The test set is used **after training is complete** to evaluate the final model’s performance.
* It acts as completely new data to assess how well the model is expected to perform in real-world scenarios.
* Usually around 10-20% of the data.



### Why is this split important?

* **Training Set:** Without it, the model has no data to learn from.
* **Validation Set:** Helps you fine-tune the model and choose the best parameters to improve accuracy while avoiding overfitting (where the model works well on training data but poorly on new data).
* **Test Set:** Provides an unbiased evaluation of the model’s real-world performance on data it has never seen.


### Example:

Imagine you're building a model to detect spam emails:

* **Training set:** The model reads thousands of emails labeled as "spam" or "not spam" and learns patterns.
* **Validation set:** During training, you check how well the model performs on this separate set to tweak the model for better accuracy.
* **Test set:** Once trained, the model is tested on new emails to see if it correctly identifies spam in the real world.



Q7: How can unsupervised learning be used in anomaly detection?

ans.


### Anomaly Detection?

Anomaly detection is the process of identifying data points that **deviate significantly** from the normal pattern or behavior in a dataset. These unusual points are called **anomalies** or **outliers**.



### Using Unsupervised Learning for Anomaly Detection:

* In many real-world scenarios, **labeled data for anomalies is rare or unavailable**, so supervised learning can’t be used effectively.
* **Unsupervised learning algorithms** analyze the structure of the **normal (majority) data** and learn what “typical” looks like.
* Once the model understands the normal patterns, it can flag data points that **don’t fit well** as anomalies.


### How does it work?

1. **Model learns the normal data distribution or clusters** without any labels.
2. Data points that are far away from these clusters or that don’t conform to the learned pattern are considered anomalies.



### Examples of Unsupervised Algorithms for Anomaly Detection:

* **Clustering (e.g., K-Means):** Points that don’t belong clearly to any cluster can be anomalies.
* **Isolation Forest:** Builds random trees to isolate data points; anomalies get isolated faster because they are different.
* **Autoencoders (Deep Learning):** Neural networks learn to compress and reconstruct data; data points that reconstruct poorly are potential anomalies.



### Real-world Use Cases:

* **Fraud Detection:** Spotting unusual transactions in banking that differ from regular spending patterns.
* **Network Security:** Detecting unusual activity or intrusions in network traffic.
* **Manufacturing:** Identifying defective products based on sensor data that don’t match the usual operating conditions.



Q8: List down some commonly used supervised learning algorithms and unsupervised learning algorithms.

ans.

### Commonly Used **Supervised Learning** Algorithms:

1. **Linear Regression**

   * Used for predicting continuous values (e.g., house prices).
2. **Logistic Regression**

   * Used for binary classification (e.g., spam vs. not spam).
3. **Decision Trees**

   * Tree-based model used for classification and regression.
4. **Random Forest**

   * Ensemble of decision trees, improves accuracy and reduces overfitting.
5. **Support Vector Machines (SVM)**

   * Classifies data by finding the optimal boundary between classes.
6. **K-Nearest Neighbors (KNN)**

   * Classifies based on the majority class among the nearest neighbors.
7. **Naive Bayes**

   * Probabilistic classifier based on Bayes’ theorem, useful for text classification.
8. **Gradient Boosting Machines (e.g., XGBoost, LightGBM)**

   * Ensemble methods that build models sequentially to correct errors of previous models.
9. **Neural Networks**

   * Used for complex tasks like image recognition and natural language processing.



### Commonly Used **Unsupervised Learning** Algorithms:

1. **K-Means Clustering**

   * Groups data into k clusters based on feature similarity.
2. **Hierarchical Clustering**

   * Builds a tree of clusters (dendrogram) based on data similarity.
3. **DBSCAN (Density-Based Spatial Clustering)**

   * Clusters based on data density, good for detecting noise and clusters of different shapes.
4. **Principal Component Analysis (PCA)**

   * Reduces dimensionality of data by transforming it into principal components.
5. **Isolation Forest**

   * Detects anomalies by isolating outliers in data.
6. **Autoencoders (Deep Learning)**

   * Neural networks that learn data representation and are used for anomaly detection and dimensionality reduction.
7. **Gaussian Mixture Models (GMM)**

   * Models data as a mixture of several Gaussian distributions, used for clustering.
