Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.express as px

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Scatter plot: Age vs Fare
fig = px.scatter(titanic, x='age', y='fare', title="Scatter Plot of Age vs Fare")
fig.show()
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
# Load tips dataset
tips = px.data.tips()

# Box plot
fig = px.box(tips, x='day', y='total_bill', color='sex', title="Box Plot of Total Bill by Day and Sex")
fig.show()
```

Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

```python
# Histogram with additional parameters
fig = px.histogram(
    tips,
    x="sex",
    y="total_bill",
    color="day",
    pattern_shape="smoker",
    histfunc="avg",
    title="Histogram of Total Bill by Sex, Day, and Smoker Status"
)
fig.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.


```python
import plotly.express as px

# Load the iris dataset from Plotly
iris = px.data.iris()

# Scatter matrix plot using selected dimensions and species for color
fig = px.scatter_matrix(
    iris,
    dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
    color="species",
    title="Scatter Matrix of Iris Dataset"
)

fig.show()
```

Q5. What is Distplot? Using Plotly express, plot a distplot.

ans. A distplot (distribution plot) is used to visualize the distribution of a numeric variable. It typically combines a histogram and a Kernel Density Estimate (KDE) to show how data is spread across values.

However, Plotly Express does not have a built-in distplot function. Instead, you can use plotly.figure_factory.create_distplot() for similar functionality.



```python
import plotly.figure_factory as ff
import numpy as np

# Generate random data
x = np.random.randn(500)

# Create distplot
fig = ff.create_distplot(
    [x],                 # Data (list of lists)
    group_labels=["Data"],  # Legend label
    show_hist=True,         # Show histogram
    show_rug=True           # Show rug plot
)

fig.update_layout(title="Distplot of Randomly Generated Data")
fig.show()
```
