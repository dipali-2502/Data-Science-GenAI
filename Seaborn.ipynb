Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ans.
Line Plot (sns.lineplot)
Used to display trends over time or continuous variables.

Box Plot (sns.boxplot)
Used to visualize the distribution, median, quartiles, and outliers of numerical data.

Histogram (sns.histplot)
Used to show the frequency distribution of a single variable.

Pair Plot (sns.pairplot)
Used to visualize relationships between all pairs of variables in a dataset.

Heatmap (sns.heatmap)
Used to display data in a matrix format with color gradients representing values

Que 2: Load the "fmri" dataset using Seaborn and plot a line plot using timepoint vs signal for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load fmri dataset
fmri = sns.load_dataset('fmri')

# Line plot
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')
plt.title('FMRI Signal over Time')
plt.show()
```

Que 3: Load the "titanic" dataset and plot two box plots using x='pclass', y='age' and y='fare'.


```python
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(12, 5))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age distribution by Passenger Class')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare distribution by Passenger Class')

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset to plot a histogram for the 'price' column with hue='cut'.


```python
diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title('Histogram of Diamond Prices by Cut')
plt.show()
```

Que 5: Use the "iris" dataset to plot a pair plot with hue='species'.


```python
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

que 6: Use the "flights" dataset to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

# Correct pivot call with keywords
flights_pivot = flights.pivot(index='year', columns='month', values='passengers')

sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Flights Heatmap: Passengers per Month and Year')
plt.show()
```
